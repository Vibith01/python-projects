```python
def player_choice():
    choice = input("Odd\Even :")
    global odd
    global even
    
    if choice == 'odd':
        odd = 'player' 
        even = 'computer'
        print('Player will be odd , Computer will be even')
    elif choice == 'even':
        even = 'player'
        odd = 'computer'
        print('Player will be even , Computer will be odd')
    return odd,even     
```

```python
def player_batting():
    total = 0
    p = input("Enter ur bat :")
    while p not in '123456':
        p = input("Invalid Input.Enter a num between 1-6 :")
    p = int(p)
    c = random.randint(1,7)
    print(f"Computer bowls: {c}")
    if p == c:
        print("Out")
    else:
        total += p
        print(f"Your current score: {total}")
    while p != c:
        p = input("Enter ur bat :")
        while p not in '123456':
            p = input("Invalid Input.Enter a num between 1-6 :")
        p = int(p)
        c = random.randint(1,7)
        print(f"Computer bowls: {c}")
        total += p
        print(f"Your current score: {total}")
    if p == c:
        print(f"player's current score {total}.computer's turn")
    return total
```
